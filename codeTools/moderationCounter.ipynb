```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the datasets
a0 = pd.read_csv('/Users/george/Desktop/Uni/2023-2024/Thesis/Final Thesis Collection/testCSV/nomoderation2.csv')
a1 = pd.read_csv('/Users/george/Desktop/Uni/2023-2024/Thesis/Final Thesis Collection/testCSV/basicmoderation.csv')
a2 = pd.read_csv('/Users/george/Desktop/Uni/2023-2024/Thesis/Final Thesis Collection/testCSV/advancedmoderation.csv')
a3 = pd.read_csv('/Users/george/Desktop/Uni/2023-2024/Thesis/Final Thesis Collection/testCSV/selection.csv')
# Define the list of methods to summon an admin 
methods = ['@admin','/report', '/warn']
```

```python
# Count the number of times each method was used in A0
a0_count = {}
for method in methods:
    a0_count[method] = a0['message'].str.count(method).sum()
print(a0_count)
```

```python
# Count the number of times each method was used in A1
a1_count = {}
for method in methods:
    a1_count[method] = a1['message'].str.count(method).sum()
print(a1_count)
```

```python
# Count the number of times each method was used in A2
a2_count = {}
for method in methods:
    a2_count[method] = a2['message'].str.count(method).sum()
print(a2_count)
```

```python
# Count the number of times each method was used in A3
a3_count = {}
for method in methods:
    a3_count[method] = a3['message'].str.count(method).sum()
print(a3_count)
```

```python
# Compute the ratio of the number of times each method was used to the total number of messages in A0 
a0_ratio = {}
for method in methods:
    a0_ratio[method] = a0_count[method]/len(a0)
plt.bar(a0_ratio.keys(), a0_ratio.values(), color=['#FF6700', '#004E98', '#9A031E'])
title = 'Ratio of the number of times each method was used to the total number of messages in A0'
plt.title(title, loc = 'center', wrap = True)
plt.savefig('/Users/george/Desktop/Uni/2023-2024/Thesis/Final Thesis Collection/collectionTool/graphics/a0_openness.png')
```
